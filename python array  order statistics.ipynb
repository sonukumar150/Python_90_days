```python
#kth smallest/largest element in unsorted array
Approach
Sort the array using O(NlogN) sorting algorithm like Merge Sort, Heap
Sort, etc and return the element at index k-1 sorted array
```

```python
def kthsmallest(arr,n,k):
    arr.sort()
    return arr[k-1]
arr=[12,3,7,9,13]
n=len(arr)
k=2
print('kth smallest element is', kthsmallest(arr,n,k))
```

```python
#quickselect
#Optimization over method 1 
import sys
def kthSmallest(arr,l,r,k):
    if(k>0 and k<=r-l+1):
        pos=partition(arr,l,r)
        if(pos-l==k-1):
            return arr[pos]
        if(pos-l>k-1):
            return kthSmallest(arr,l,pos-1,k)
        return kthSmallest(arr,pos+1,r,k-pos+l-1)
    return sys.maxsize
def partition(arr,l,r):
    x=arr[r]
    i=l
    for j in range(l,r):
        if (arr[j]<=x):
            arr[i],arr[j]=arr[j],arr[i]
            i+=1
    arr[i],arr[r]=arr[r],arr[i]
    return i
arr=[12,3,5,7,4,19,26]
n=len(arr)
k=3
print("K'th smallest element is",kthSmallest(arr,0,n-1,k))
        
```

```python
#minimum product of k integers in an array of positive integers
import math
import heapq
def minProduct(arr,n,k):
    heapq.heapify(arr)
    count=0
    ans=1
    while(arr) and count<k:
        x=heapq.heappop(arr)
        ans=ans*x
        count=count+1
    return ans
arr=[198,76,544,123,154,675]
k=2
n=len(arr)
print('Minimum product is',minProduct(arr,n,k))
#time complexity:o(n*logn)
```

```python
#kth largest sum contiguous subarray
#Approach
store pre-sum of the array in a sum array 
for storing kth largest sum, use min heap in which we push contiguous
sum till we get k elements, once we have our k elements,check if the 
element is greater than the kth element inserted to min heap with 
popping out the top element else not inserted
```

```python
import heapq
def kthLargestSum(arr,n,k):
    sum=[]
    sum.append(0)
    sum.append(arr[0])
    for i in range(2,n+1):
        sum.append(sum[i-1]+arr[i-1])
    Q=[]
    heapq.heapify(Q)
    for i in range(1,n+1):
        for j in range(i,n+1):
            x=sum[j]-sum[i-1]
            if len(Q)<k:
                heapq.heappush(Q,x)
            else:
                if Q[0]<x:
                    heapq.heappop(Q)
                    heapq.heappush(Q,x)
    return Q[0]
a=[10,-10,20,-40]
n=len(a)
k=6
print(kthLargestSum(a,n,k))
```

```python
#Alternate sorting
Given array of integers, print the array in such a way that the first 
element is the first maximum and second element is first minimum 
#Approach
print first maximum, then minimum, then second maximum and so on
#time complexity O(n**2)
#Alternate approach
1) Sort input array using O(nlogn) algorithm
2) We maintain 2 pointers, one from beginning and one from the end in 
   sorted array. Alternatively print elements pointed by 2 pointers and 
   move them towards each other
```

```python
def alternateSort(arr,n):
    arr.sort()
    i=0
    j=n-1
    while(i<j):
        print(arr[j],end=" ")
        j-=1
        print(arr[i], end=" ")
        i+=1
        if(n%2!=0):
            print(arr[i])
arr=[1,12,4,6,7,10]
n=len(arr)
alternateSort(arr,n)
```

```python
#triplets with zero sum
#method1
#Approach
run 3 loops and check one by one that sum of 3 elements is zero or not
If the sum of 3 elements is zero then print elements otherwise print
not found
#Algorithm
1) Run 3 nested loops with loop counter i,j,k
2) The 3 loop will run from 0 to n-3 and second loop from i+1 to n-2 
   and 3rd loop from j+1 to n-1. The loop counter represent 3 elements
   of the triplet
3) check the sum of element at i'th, j'th, k'th is equal to zero or 
   not .If yes print the sum else continue
 #Time complexity: O(n*n*n) 
 #Auxiliary space:O(1)
```

```python
def findTriplets(arr,n):
    found=True
    for i in range(0,n-2):
        for j in range(i+1,n-1):
            for k in range(j+1,n):
                if(arr[i]+arr[j]+arr[k]==0):
                    print(arr[i],arr[j],arr[k])
                    found=True
    if(found==False):
        print("not exist")
arr=[0,-1,2,-3,1]
n=len(arr)
findTriplets(arr,n)
```

```python
#Method 2
uses the process of hashing to arrive at the result and is solved at a
lesser time of O(n*n)
Approach
It involves traversing through the array.For every element arr[i], find
a pair with sum "-arr[i]". problem reduces to pair sum and can be solved
in O(n) time using hashing
#Algorithm
1) Create a hashap to store a key value pair
2) Run a nested loop with two loops, outer loop from 0 to n-2 and the 
   inner loop from i+1 to n-1
3) Check if the sum of ith and jth element multiplied with -1 is present
   in the hashmap or not
4) If the element is present in the hashmap, print the triplet else 
   insert the j'th element in the hashmap
#Auxiliary space O(n)
```

```python
def findTriplets(arr,n):
    found=False
    for i in range(n-1):
        s=set()
        for j in range(i+1,n):
            x=-(arr[i]+arr[j])
            if x in s:
                print(x,arr[i],arr[j])
                found=True
            else:
                s.add(arr[j])
    if found==False:
        print("No Triplet Found")
arr=[0,-1,2,-3,1]
n=len(arr)
findTriplets(arr,n)
```

```python
#Method 3
Sorting to arrive at the correct result and is solved in O(n*n) time
#Approach
The idea is based on method 2
#Algorithm
1) Sort the array in ascending array
2) Transverse the array from start to end
3) For every index i, create 2 variables l=i+1 and r=n-1
4) Run a loop until l is less than r, if sum of array[l],array[r] is 
   equal to zero then print the triplet and break the loop
5) If the sum is less than zero then increment value of l, by increasing
   value of l the sum will increase as array is sorted
6) If sum is greater than 0 decrement value of r
```

```python
def findTriplets(arr,n):
    found=False
    arr.sort()
    for i in range(0,n-1):
        l=i+1
        r=n-1
        x=arr[i]
        while(l<r):
            if (x+arr[l]+arr[r]==0):
                print(x,arr[l],arr[r])
                l+=1
                r-=1
                found=True
            elif(x+arr[l]+arr[r]<0):
                l+=1
            else:
                r-=1
    if(found==False):
        print("No Triplet Found")
arr=[0,-1,2,-3,1]
n=len(arr)
findTriplets(arr,n)
```
